# Linear Regression Example 

Imports:

```python
import sys
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path

sys.path.append(str(Path().resolve().parent))

from models.LinearRegression import LinearRegression
from GLMTrainer import GLMTrainer
from optimizers.SGDOptimizer import SGDOptimizer
from loss.LSELoss import LSELoss
```

Generate some data: (You may want to use custom data here)

```python
x = np.linspace(0, 1, 10)
X = np.reshape(x, (x.shape[0], 1)) # for use in our model
y = x + np.random.rand()
x, y
```

Initialize our model, loss, optimizer:

```python
model = LinearRegression(1, 1)
optimizer = SGDOptimizer(0.05)
loss = LSELoss()

trainer = GLMTrainer(model, loss, optimizer)
```

Plot our data and line before training:

```python
fig, ax = plt.subplots()

ax.plot(x, y, 'o')
y_pred = trainer.model(X)
ax.plot(x, y_pred)
plt.show()
```

Train our model and plot results:

```python
trainer.train(X, y, 20)
```

Plot the trained model:

```python
fig, ax = plt.subplots()

ax.plot(x, y, 'o')
y_pred = trainer.model(X)
ax.plot(x, y_pred)
plt.show()
```

As you can see our modeled learned to fit the data
